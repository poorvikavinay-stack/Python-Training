Setup and basics

```python
# Q1: Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Q2: Create variables with basic data types
project_name = "Performance Analyzer"  # string
total_records = 5                      # int
passing_mark = 60.0                    # float
is_active = True                       # bool

# Q3: Print variable data types
print(type(project_name), type(total_records), type(passing_mark), type(is_active))
```

Dataset creation (pandas)

```python
# Q4: Build DataFrame
data = {
    'Roll No': [101, 102, 103, 104, 105],
    'Name': ['Aditi', 'Suresh', 'Monalisa', 'Kiran', 'Priya'],
    'Department': ['Commerce', 'Science', 'Commerce', 'Arts', 'Science'],
    'Accountancy': [88, 42, 95, 70, 55],
    'Economics': [75, 48, 92, 65, 58]
}
df = pd.DataFrame(data)

# Q5: Add Average column
df['Average'] = (df['Accountancy'] + df['Economics']) / 2
print(df)
```

Logic and functions

```python
# Q6: Use loop + if/else for result classification
results = []
for score in df['Average']:
    if score >= 60:
        results.append('Pass')
    else:
        results.append('Fail')
df['Result'] = results

# Q7-Q9: Custom Functions
def calculate_total(s1, s2):
    return s1 + s2

def assign_grade(avg):
    if avg >= 80: return 'A'
    elif avg >= 60: return 'B'
    else: return 'C'

def find_topper(data_frame):
    return data_frame.loc[data_frame['Average'].idxmax()]['Name']

df['Total'] = df.apply(lambda x: calculate_total(x['Accountancy'], x['Economics']), axis=1)
df['Grade'] = df['Average'].apply(assign_grade)
print(f"Top Student: {find_topper(df)}")
```

Object-oriented programming

```python
# Q10-Q13: OOP Class
class StudentReport:
    def __init__(self, dframe):
        self.dframe = dframe

    def get_summary(self):
        p_count = len(self.dframe[self.dframe['Result'] == 'Pass'])
        f_count = len(self.dframe[self.dframe['Result'] == 'Fail'])
        print(f"Pass: {p_count}, Fail: {f_count}")

    def dept_stats(self):
        print("\nAvg by Dept:\n", self.dframe.groupby('Department')['Average'].mean())

# Instantiate and call
report_tool = StudentReport(df)
report_tool.get_summary()
report_tool.dept_stats()
```

Numpy and seaborn

```python
# Q14-Q15: NumPy
avg_array = df['Average'].to_numpy()
print(f"Mean: {np.mean(avg_array):.2f}, Std Dev: {np.std(avg_array):.2f}")

# Q16-Q17: Visualizations
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.barplot(x='Department', y='Average', data=df, palette='viridis')
plt.subplot(1,2,2)
sns.countplot(x='Result', data=df, palette='magma')
plt.show()
```

Final Export

```python
# Q18: Filter pass students
pass_df = df[df['Result'] == 'Pass']

# Q19: Sort by Average descending
sorted_df = df.sort_values(by='Average', ascending=False)

# Q20: Save final report to CSV
sorted_df.to_csv('student_report.csv', index=False)
print("file saved successfully")

# Show final view
sorted_df[['Roll No', 'Name', 'Average', 'Grade', 'Result']]
```
