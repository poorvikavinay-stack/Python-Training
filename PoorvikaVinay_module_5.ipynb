Inheritance

```python
# TODO: Complete inheritance example
class Person:
    def greet(self):
        print("Hello! Welcome to the college portal.")

class Teacher(Person):
    def __init__(self, subject):
        self.subject = subject

    def show_subject(self):
        print(f"The teacher is assigned to the subject: {self.subject}")

# TODO: Create object of Teacher and call both methods
my_teacher = Teacher("Business Economics")
my_teacher.greet()
my_teacher.show_subject()
```

Encapsulation

```python
# TODO: Complete encapsulation example
class BankAccount:
    def __init__(self, initial_balance):
        self.__balance = initial_balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Successfully deposited: {amount}")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Successfully withdrawn: {amount}")
        else:
            print("Transaction failed: Insufficient balance.")

    def get_balance(self):
        return self.__balance

# TODO: Test deposit and withdraw
my_account = BankAccount(5000)
my_account.deposit(1500)
my_account.withdraw(2000)
print(f"Current Balance: {my_account.get_balance()}")
```

Polymorphism

```python
# TODO: Complete polymorphism example
class Dog:
    def sound(self):
        print("Dog says: Woof Woof!")

class Cat:
    def sound(self):
        print("Cat says: Meow!")

# Loop to call sound() for both objects
animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```
