```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set aesthetics
sns.set_theme(style="darkgrid")
print("Libraries imported successfully!")
```

```python
# Create a list of marks
marks = [82, 74, 91, 65, 88, 95, 78, 84]

# Create NumPy array
marks_array = np.array(marks)

print(f"NumPy Array: {marks_array}")
print(f"Average (Mean) Marks: {np.mean(marks_array):.2f}")
print(f"Highest Mark: {np.max(marks_array)}")
print(f"Lowest Mark: {np.min(marks_array)}")
```

```python
# Create DataFrame
data = {
    'Name': ['Ahmed', 'Sabeer', 'Zain', 'Sara', 'Ali', 'Hiba'],
    'Marks': [85, 92, 78, 88, 95, 70],
    'Department': ['CS', 'Math', 'CS', 'Physics', 'Math', 'Physics']
}
df = pd.DataFrame(data)

print("--- First 5 Rows ---")
print(df.head())

print("\n--- Summary Statistics ---")
print(df.describe())

print("\n--- Average Marks by Department ---")
print(df.groupby('Department')['Marks'].mean())
```

```python
# Create a figure with two subplots
plt.figure(figsize=(12, 5))

# 1. Bar Plot
plt.subplot(1, 2, 1)
sns.barplot(data=df, x='Department', y='Marks', palette='viridis')
plt.title('Average Marks by Department')

# 2. Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Department', y='Marks')
plt.title('Distribution of Marks by Department')

plt.tight_layout()
plt.show()
```

```python
# 1. Create DataFrame with missing values
dirty_data = {
    'Name': ['Student A', 'Student B', 'Student C', 'Student D', 'Student E'],
    'Marks': [80, np.nan, 90, 60, np.nan] # np.nan represents missing value
}
df2 = pd.DataFrame(dirty_data)

# 2. Find missing values
print("Missing values per column:\n", df2.isna().sum())

# 3. Fill missing marks with class average
class_avg = df2['Marks'].mean()
df2['Marks'] = df2['Marks'].fillna(class_avg)

# 4. Add Result column
df2['Result'] = df2['Marks'].apply(lambda x: 'Pass' if x >= 75 else 'Needs Improvement')

# 5. Filter Pass students and 6. Sort by Marks
pass_students = df2[df2['Result'] == 'Pass'].sort_values(by='Marks', ascending=False)

print("\n--- Processed & Filtered Data ---")
print(pass_students)
```

1. Difference between NumPy array and Python list?
NumPy arrays are more memory-efficient and faster for mathematical operations because they store data in a contiguous block of memory and require all elements to be of the same type (homogenous).
2. Why is Pandas useful?
Pandas allows for easy manipulation of tabular data. It provides powerful tools for reading CSV files, handling missing data, and performing SQL-like operations (joining, filtering, grouping).
3. What does Seaborn help us do?
Seaborn simplifies the creation of complex statistical visualizations. It integrates perfectly with Pandas DataFrames and automatically handles professional styling and color palettes.
